```python
from pgmpy.models import DiscreteBayesianNetwork
from pgmpy.factors.discrete import TabularCPD
from pgmpy.inference import VariableElimination

# Step 1: Define model structure (ADD ALL VARIABLES HERE)
model = DiscreteBayesianNetwork([
    ('Income', 'LoanDefault'),
    ('Credit', 'LoanDefault')   # ✅ Credit added here
])

# Step 2: Define CPDs

# Income CPD
cpd_income = TabularCPD(
    variable='Income',
    variable_card=2,
    values=[[0.6], [0.4]]
)

# Credit CPD
cpd_credit = TabularCPD(
    variable='Credit',
    variable_card=2,
    values=[[0.7], [0.3]]
)

# LoanDefault CPD (depends on Income & Credit)
cpd_default = TabularCPD(
    variable='LoanDefault',
    variable_card=2,
    values=[
        [0.9, 0.6, 0.7, 0.1],   # No Default
        [0.1, 0.4, 0.3, 0.9]    # Default
    ],
    evidence=['Income', 'Credit'],
    evidence_card=[2, 2]
)

# Step 3: Add CPDs to model
model.add_cpds(cpd_income, cpd_credit, cpd_default)

# Step 4: Check model
print(model.check_model())  # Should print True

# Step 5: Inference
inference = VariableElimination(model)

result = inference.query(variables=['LoanDefault'], evidence={'Income': 1, 'Credit': 0})
print(result)
```
